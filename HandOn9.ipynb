```python
class Link():
    def __init__(self, key, data):
        self.next = None
        self.previous = None
        self.key = key
        self.data = data
    def __str__(self):
        return f'{self.key, self.data}'
        
class LinkList():
    def __init__(self):
        self.head = None
        self.tail = None
        self.size = 0
    def insertFro(self, key, data):
        link = Link(key, data)
        if self.head == None:
            self.head = link
            self.tail = link
        else:
            link.next = self.head
            self.head.previous = link
            self.head = link
        self.size += 1
    def search(self, x):        # search for key = x, return data
        curr = self.head
        look = x
        while(curr):
            if curr.key == look:
                return curr.data
            curr = curr.next
        return -191919
    def delist(self, x):                #           ******t look for key, delete link
        look = x
        if self.head.key == look:
            if self.head.next:
                self.head = self.head.next
                self.head.previous = None
            else:
                self.head = None
            self.size -= 1
            return
        curr = self.head
        while(curr):
            prev = curr
            curr = curr.next
            if curr and curr.key == look:
                if curr.next == None:
                    prev.next = None
                else:
                    prev.next = curr.next
                    curr.next.previous = prev
                self.size += -1
                return
        return -1
    def printLi(self):
        curr = self.head
        while(curr):
            print(curr)
            curr = curr.next

```

```python
import math
class HashT():
    def __init__(self, m, fun):
        self.capacity = m
        self.size = 0
        self.fun = fun
        self.hashT = np.empty(self.capacity, dtype=object)
        for i in range(self.capacity):
            self.hashT[i] = LinkList()
    def hashit(self, key):
            if self.fun == 1:
                return key % self.capacity
            elif self.fun == 2:
                A = (math.sqrt(5)- 1)/2
                return math.floor(self.capacity * ((key * A) % 1))
    def rehash(self, newCap): 
        oldie = self.hashT
        self.capacity = newCap
        self.hashT = np.empty(self.capacity, dtype=object)
        for i in range(self.capacity):
            self.hashT[i] = LinkList()
        self.size = 0
        for ind in range(len(oldie)):
            lis = oldie[ind]
            curr = lis.head
            while curr:
                self.insert(curr.key, curr.data)
                curr = curr.next
    def insert(self, key, data):
        if self.size >= self.capacity:        
            self.rehash( self.capacity * 2)
        ind = self.hashit(key)
        lis = self.hashT[ind]
        lis.insertFro(key, data)
        self.size += 1
    def find(self, key):
        ind = self.hashit(key)
        found = self.hashT[ind].search(key)
        if found == -191919:
            return ('not found')
        else:
            return found
    def dele(self, key):
        ind = self.hashit(key)
        d = self.hashT[ind].delist(key)
        if d == -1:
            print("key ", key, " not found")
            return
        self.size -= 1
        if self.size <= self.capacity // 4:
            self.rehash( self.capacity // 2)
    def pri(self):
        print("*** size ", self.size , " capacity ", self.capacity, "  ***")
        for i in range(self.capacity):
            print("ind: ",i, " length of ", self.hashT[i].size)
            self.hashT[i].printLi()
```

```python
ha = HashT(5, 1)
ha.insert(1, 1)
ha.insert(11, 22)
ha.insert(69, 2)
ha.insert(16, 321)
ha.pri()
print("<> inserting 5th element <>")
ha.insert(121, 922)
ha.dele(11)
ha.insert(123, 777)
ha.pri()
print("<>  inserting 6th element <>")

ha.insert(62, 42)
ha.pri()
```
