```python

def height(node):
    if node is None:
        return -1
    else:
        return node.height
    
class Node():
    def __init__(self, key, parent = None):
        self.parent = parent        
        self.key = key
        self.left = None
        self.right = None
        self.height = None        
    def up_height(self):
        self.height = max(height(self.right), height(self.left)) + 1
    
    def rebalance(self, tree):
        if self.key is None:
            return
        self.up_height()
        if height(self.left) > (height(self.right) + 1):    
            if (height(self.left.right) > height(self.left.left)):
                self.left.leftRot(tree)
            self.rightRot(tree)
        elif height(self.right) > (height(self.left) + 1):
            if  (height(self.right.left) > height(self.right.right)):
                self.right.rightRot(tree)
            self.leftRot(tree)
        if(self.parent):
            self.parent.rebalance(tree)
    
    def rightRot(self, tree):
        s = self
        l = self.left
        
        l.parent = s.parent
        
        if s.parent is None:
            tree.root = l
        if s.parent.left is s:
            s.parent.left = l
        elif s.parent.right is s:
            s.parent.right = l
        s.parent = l
        s.left = l.right
        if s.left:
            s.left.parent = s
        l.right = s
        s.up_height()
        l.up_height()
        
    def leftRot(self, tree):
        s = self
        r = self.right
        r.parent = s.parent
        if s.parent is None:
            tree.root = r
        elif s.parent.left is s:
            s.parent.left = r
        elif s.parent.right is s:
            s.parent.right = r
        s.parent = r
        s.right = r.left
        if s.right:
            s.right.parent = s
        r.left = s
        s.up_height()
        r.up_height()
    
        
    def __str__(self):
        return str(self.key)


class AVL():
    def __init__(self):
        self.root = None
    def insert(self, key):
        node = Node(key)
        if self.root is None:
            self.root = node
            return
        else:
            par = None
            curr = self.root
            while curr:
                par = curr
                if node.key < curr.key:
                    if not curr.left:
                        curr.left = node
                        curr.left.parent = curr
                        curr.left.rebalance(self)
                        return
                    curr = curr.left
                else:
                    if not curr.right:
                        curr.right = node
                        curr.right.parent = curr
                        curr.right.rebalance(self)
                        return
                    curr = curr.right
    def remove(self, k):        
        d = self.search(k)
        self = self.delete(d)
    def delete(self, node):
        if node is None:
            return
        if node.left == None and node.right == None:
            if node.parent == None:
                self.root = None
            elif node.parent.left == node:
                node.parent.left = None
            elif node.parent.right == node:
                node.parent.right = None
            node.rebalance(self)                       ###n.par?
            return 
        if node.left and node.right:
            success = self.succTree(node)
            node.key = success.key
            self.delete(success)
            node.rebalance(self)
            return
        elif node.left:
            if node == self.root:
                self.root, node.left.parent = node.left, None
                return
            if node.parent.right == node:
                node.parent.right, node.left.parent = node.left, node.parent #      
            elif node.parent.left == node:
                node.parent.left, node.left.parent = node.left, node.parent
            node.rebalance(self)
        elif node.right:
            if node == self.root:
                self.root, node.right.parent = node.right, None
                return
            if node.parent.right == node:
                node.parent.right, node.right.parent = node.right, node.parent      
            elif node.parent.left == node:
                node.parent.left, node.right.parent = node.right, node.parent 
            node.rebalance(self)
    def treeMin(self, node):
        curr = node
        while curr.left:
            curr = curr.left
        return curr
    def treeMax(self, node):
        curr = node
        while curr.right:
            curr = curr.right
        return curr
    def succTree(self, node):
        if node.right is not None:
            return self.treeMin(node.right)
        succ = node.parent
        while succ and node is succ.right:
            node = succ
            succ = succ.parent
        return succ
    def succ(self, k):
        k_node = self.search(k)
        success = self.succTree(k_node)
        print("the successor is " , success.key)
    def predTree(self, node):
            if node.left is not None:
                return self.treeMax(node.left)
            pred = node.parent
            while pred and node is pred.left:
                node = pred
                pred = pred.parent
            return pred
    def pred(self, k):
        k_node = self.search(k)
        pred = self.predTree(k_node)
        print("the predecessor is " , pred.key)
    def inoPrint(self, node):
        if node is not None:
            self.inoPrint(node.left)
            p = None
            l = None
            r = None
            lh = -1
            rh = -1
            if node.parent:
                p = node.parent
            if node.left:
                l = node.left
                lh = l.height
            if node.right:
                r = node.right
                rh = r.height
            print("node: " ,node, " h ",node.height, "  p : ", p, "  l: ", l, " lheight: ", lh, "  r: ", r, " rheight ", rh)
            self.inoPrint(node.right)
    def search(self, k):
        node = self.root
        while (node) and (k != node.key):
            if k < node.key:
                node = node.left
            else:
                node = node.right
        if node and node.key == k:
            print("node ", node, "found.")
            return node
        else:
            print("node not found")
            return
```
