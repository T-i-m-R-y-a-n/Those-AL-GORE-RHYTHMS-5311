```python
import copy
class Node():
    def __init__(self, key):  # data not used
        self.key = key
        # self.data = data (not used)
        self.parent = None
        self.left = None
        self.right = None
    def __str__(self):
        return str(self.key)
class BST():
    def __init__(self):
        self.root = None
        # self.size = 0
    def inoPrint(self, node):
        if node is not None:
            self.inoPrint(node.left)
            p = None
            l = None
            r = None
            if node.parent:
                p = node.parent
            if node.left:
                l = node.left
            if node.right:
                r = node.right
            print("node: " ,node, "  p : ", p, "  l: ", l,"  r: ", r)
            self.inoPrint(node.right)
    def search(self, k):
        """  recursive
        if (node == None) or (node.key == k):
            return node
        if k < self.key:
            return search(node.left, k)
        else:
            return search(node.right, k)
        """
        node = self.root
        while (node) and (k != node.key):
            if k < node.key:
                node = node.left
            else:
                node = node.right
        if node and node.key == k:
            print("node ", node, "found.")
            return node
        else:
            print("node not found")
            return
    def insert(self, key):
        node = Node(key)
        y = None
        if self.root is None:
            self.root = node
            return
        else:
            par = None
            curr = self.root
            while curr:
                par = curr
                if node.key < curr.key:
                    if not curr.left:
                        curr.left = node
                        curr.left.parent = curr
                        return
                    curr = curr.left
                else:
                    if not curr.right:
                        curr.right = node
                        curr.right.parent = curr
                        return
                    curr = curr.right
    def remove(self, k):        
        d = self.search(k)
        self = self.delete(d)
    def delete(self, node):
        if node is None:
            return
        if node.left == None and node.right == None:
            if node.parent == None:
                self.root = None
            elif node.parent.left == node:
                node.parent.left = None
            elif node.parent.right == node:
                node.parent.right = None
            return 
        if node.left and node.right:
            success = self.succTree(node)
            node.key = success.key
            self.delete(success)
            return
        elif node.left:
            if node == self.root:
                self.root, node.left.parent = node.left, None
                return
            if node.parent.right == node:
                node.parent.right, node.left.parent = node.left, node.parent #      
            elif node.parent.left == node:
                node.parent.left, node.left.parent = node.left, node.parent
        elif node.right:
            if node == self.root:
                self.root, node.right.parent = node.right, None
                return
            if node.parent.right == node:
                node.parent.right, node.right.parent = node.right, node.parent      
            elif node.parent.left == node:
                node.parent.left, node.right.parent = node.right, node.parent #
    def treeMin(self, node):
        curr = node
        while curr.left:
            curr = curr.left
        return curr
    def treeMax(self, node):
        curr = node
        while curr.right:
            curr = curr.right
        return curr
    def succTree(self, node):
        if node.right is not None:
            return self.treeMin(node.right)
        succ = node.parent
        while succ and node is succ.right:
            node = succ
            succ = succ.parent
        return succ
    def succ(self, k):
        k_node = self.search(k)
        success = self.succTree(k_node)
        print("the successor is " , success.key)
    def predTree(self, node):
            if node.left is not None:
                return self.treeMax(node.left)
            pred = node.parent
            while pred and node is pred.left:
                node = pred
                pred = pred.parent
            return pred
    def pred(self, k):
        k_node = self.search(k)
        pred = self.predTree(k_node)
        print("the predecessor is " , pred.key)
```
